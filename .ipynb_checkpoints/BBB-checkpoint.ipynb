# eMMC Testing #

[Revisiting Storage for Smartphones](http://static.usenix.org/events/fast/tech/full_papers/Kim.pdf) gives a good overview of different performance characteristics on Android phones.  They find that random writes can be abysmal (0.02 MB/s or less). Also Kingston consistently performed the worst, and Transcent performs the best. They basically analyzed how different android apps performed with different eMMC memories, and even did some observation on web caching, which is useful.

```python
rs = 'read_status_'
r = 'r_'
w = 'w_'
s = 'submission_latency_'
c = 'completion_latency_'
clp = 'comp_lat_%_'
tl = 'total_latency_'
bw = 'bw_'
ws = 'write_status_'
iodepth = 'io_depth_%_'

mmms = [
'min', 'max', 'mean', 'stddev'
]

mmms_bw = [
'min', 'max', 'aggr_%_total', 'mean', 'stddev'
]


sumsMetrics = [
    'total_io',
    'bw',
    'iops',
    'runtime'
]

fioOut = [
    'terse version',
    'fio version',
    'jobname',
    'groupid',
    'error'
]

cpuUsage = [
    'user',
    'system',
    'context_switches',
    'major_pfs',
    'minor_pfs'
]

fioOut.extend([rs + stat for stat in sumsMetrics])
fioOut.extend([r + s + stat for stat in mmms])
fioOut.extend([r + c + stat for stat in mmms])
fioOut.extend([r + clp + str(i) for i in range(0,20)])
fioOut.extend([r + tl + stat for stat in mmms])
fioOut.extend([r + bw + stat for stat in mmms_bw])
fioOut.extend([ws + stat for stat in sumsMetrics])
fioOut.extend([w + s + stat for stat in mmms])
fioOut.extend([w + c + stat for stat in mmms])
fioOut.extend([w + clp + str(i) for i in range(0,20)])
fioOut.extend([w + tl + stat for stat in mmms])
fioOut.extend([w + bw + stat for stat in mmms_bw])
fioOut.extend([usage for usage in cpuUsage])
fioOut.extend([iodepth + str(i) for i in ['<=1',2,4,8,16,'>=64' ]])


highlights = [rs + stat for stat in sumsMetrics]

highlights.extend([ws + stat for stat in sumsMetrics])



fioOutTerse = {}
for highlight in highlights:
    fioOutTerse[highlight] = fioOut.index(highlight)
    
```

```python
def realVal(v):
    try:
        return v/4000
    except:
        return int(v)/4000

def kbToMb(x):
    return int(x)/1000
```

```python
def printStats(f):
    for line in f:
        parsedLine = line.split(';')
        print ""
        key = "read_status_total_io"
        val = 'read_status_bw'
        for tag in fioOutTerse:
            print "%s : %s" % (tag, parsedLine[fioOutTerse[tag]])
        print '\n'
        

#reqData is an array of strings of the data we'd like to get. 
# The result is a dictionary of the lenth of reqData of arrays
def getData(reqData, f):
    resData = {}
    for datum in reqData:
        assert datum in fioOut
        resData[datum] = []
        
    for line in f:
        parsedLine = line.split(';')
        for datum in resData:
            resData[datum].append(parsedLine[fioOut.index(datum)])
    return resData

```

```python
# plot (potentially) multiple y values versus an x value

def makePlot(xVal, yVals, labels = None, xlabel=None, ylabel=None, title=None):
    try: plt
    except:
        %matplotlib inline
        import matplotlib.pyplot as plt
    try: np
    except:
        import numpy as np

    if type(yVals[0]) is list:
        for yVal in yVals:
            label = None
            try:
                label = labels[yVals.index(yVal)]
            except:
                label = None
            plt.plot(xVal, yVal, label=label)
    else:
        plt.plot(xVal, yVals)
        
    plt.legend()
    if xlabel: plt.xlabel(xlabel)
    if ylabel: plt.ylabel(ylabel)
    if title:  plt.title(title)
    plt.tight_layout()

```

##Random Reads vs Cache Size,4KB and 8KB Chunks##

```python

trace3 = open('./results/randread4k1')
trace4 = open('./results/randread8k1')
readbw4k = getData(['read_status_bw'], trace3)['read_status_bw']
readbw8k = getData(['read_status_bw'], trace4)['read_status_bw']

makePlot([200,400,600,800,1000,1200,1400,1600], [readbw4k,readbw8k],['4k units', '8k units'], xlabel="Cache size (MB)", ylabel="Read BW (KB/s)", title="Random Reads")
```
